```python
import pandas as pd

data = pd.read_csv("annotated_prompts_with_intent.csv")
data.head()
```

```python
X = data['Prompt']
y = data['Intent']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_vectorized = vectorizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Example: user prompt test
test_prompt = ["What are common emotional symptoms of PMS and how to cope?"]
test_vector = vectorizer.transform(test_prompt)
print("Predicted Intent:", model.predict(test_vector)[0])
```

```python
#Clarification Ability Testing
# Test model's clarification ability
clarification_prompts = [
    "I feel weird during periods",
    "Something's wrong but I don't know what",
    "Can you help me with this feeling?",
    "Pain but not physical"
]

# Vectorize clarification prompts
clarification_vectorized = vectorizer.transform(clarification_prompts)

# Predict
clarification_preds = model.predict(clarification_vectorized)

# Show results
for prompt, pred in zip(clarification_prompts, clarification_preds):
    print(f"Prompt: {prompt} → Predicted Intent: {pred}")
```

```python
#Symptom Recognition Testing
# Select some rows where symptom is clearly mentioned
symptom_test_data = data[data['Symptom'].notnull()].sample(5)

# Vectorize and predict
symptom_vectorized = vectorizer.transform(symptom_test_data['Prompt'])
symptom_preds = model.predict(symptom_vectorized)

# Compare with actual intents
for i in range(len(symptom_test_data)):
    print(f"Prompt: {symptom_test_data.iloc[i]['Prompt']}")
    print(f"→ Predicted Intent: {symptom_preds[i]}")
    print(f"→ Actual Intent: {symptom_test_data.iloc[i]['Intent']}\n")
```

```python
# User Feedback Module
# Simulated user feedback
feedback_data = [
    {"prompt": "What should I do for stress?", "predicted_intent": "mental_wellness", "feedback": "Helpful"},
    {"prompt": "I'm feeling pain and confused", "predicted_intent": "diet_advice", "feedback": "Not helpful"},
    {"prompt": "I'm scared and anxious", "predicted_intent": "mental_wellness", "feedback": "Helpful"}
]

# Display feedback summary
for fb in feedback_data:
    print(f"Prompt: {fb['prompt']}")
    print(f"→ Predicted Intent: {fb['predicted_intent']}")
    print(f"→ User Feedback: {fb['feedback']}\n")
```
